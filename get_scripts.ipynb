# Get links

```python
import urllib.request as ur
import pandas as pd
from bs4 import BeautifulSoup
import re
```

```python
mainpage = ur.urlopen("http://www.imsdb.com/TV/Seinfeld.html")
mainhtml = str(mainpage.read())
```

```python
def getlinks(html):
    links=[]
    equote = 0
    while len(links) < 300:
        slink = mainhtml.find('a href=',equote) # find the first link after the equote
        squote = mainhtml.find('"',slink) # find the first " after the start of the link
        equote = mainhtml.find('"',squote + 1) # find the next " after the previous "
        links.append(mainhtml[squote+1:equote])
    return links
```

```python
links = getlinks(mainhtml)[64:64+176]
# the list of episodes starts at link 64 and there are only 176 episodes of Seinfeld
```

# Clean Links

```python
for i in range(len(links)):
    links[i] = "https://www.imsdb.com/t" + links[i][5:]
    links[i] = links[i][:-12] + links[i][-5:]
    links[i] = ''.join(links[i].split('- '))
    links[i] = '-'.join(links[i].split(' '))
```
