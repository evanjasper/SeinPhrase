```python
stgdir = []
dialog = []

for i in scripts['Dialogue']:
    stgdir.append('')
    dialog.append('')
    dlogcopy = str(i)
    sd = dlogcopy.find('(')
    if sd == -1:
        dialog[-1] += dlogcopy
    while sd != -1:
        sd = dlogcopy.find('(')
        ed = dlogcopy.find(')',sd+1)
        stgdir[-1] += dlogcopy[sd+1:ed]
        dialog[-1] += dlogcopy[:sd]
        dlogcopy = dlogcopy[ed+2:]
```
